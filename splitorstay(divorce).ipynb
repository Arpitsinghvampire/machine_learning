```python
#now we need to store them into the feature vectors 

# now let's make them into the feature vector
features = []
for i in range(len(datas)):
    vectors = []
    for col in datas.columns: 
        if col!='Class':# Iterate over column names directly
            vectors.append(datas[col][i])  # Access elements using column names
    features.append(vectors)
# from here we get the feature vector
```

```python
print(features[0]) #the last one is the class.
```

```python
#now lets prepare the x_train and the x_test 
x_train=[]
x_test=[]


for i in range(100):
    x_train.append(features[i])
    
for i in range(100,170):
    x_test.append(features[i])
#now we have prepared the x_train and the x_test

```

```python
y_train=[]
y_test=[]

for i in range(100):
    if datas['Class'][i]==1:
        y_train.append([0,1])
    if datas['Class'][i]==0:
        y_train.append([1,0])
        
for i in range(100,170):
    if datas['Class'][i]==1:
        y_test.append([0,1])
    if datas['Class'][i]==0:
        y_test.append([1,0])
```

```python
#lets typecast them into np array
import numpy as np
x_train=np.asarray(x_train)
x_test=np.asarray(x_test)

y_train=np.asarray(y_train)
y_test=np.asarray(y_test)
```

```python
print(x_train[0].shape)
```

```python
#now lets prepare the model
import tensorflow 
from tensorflow import keras 
from keras.layers import Dense,Input,BatchNormalization,Dropout

```

```python
input1=Input((54,))
dense_1=Dense(128,activation='relu')(input1)
dense_2=Dense(128,activation='relu')(dense_1)
batchnormal=BatchNormalization()(dense_2)
dense_3=Dense(64,activation='relu')(batchnormal)
drop=Dropout(0.2)(dense_3)
dense_4=Dense(64,activation='relu')(drop)

dense_5=Dense(64,activation='tanh')(dense_4)

dropo=Dropout(0.1)(dense_5)
dense_6=Dense(64,activation='relu')(dropo)

batch=BatchNormalization()(dense_6)
dense_7=Dense(32,activation='relu')(batch)

dro=Dropout(0.1)(dense_7)
dense_8=Dense(16,activation='relu')(dro)
batchNor=BatchNormalization()(dense_8)

dense_9=Dense(8,activation='tanh')(batchNor)

dropout=Dropout(0.3)(dense_9)

dense_10=Dense(2,activation='softmax')(dropout)
```

```python
#now lets wrap the layers into one
from keras import Model
from keras.utils import plot_model 

model=Model(inputs=input1,outputs=dense_10)
```

```python
#lets print the summary. of the model 
model.summary()
```

```python
#now lets plot the model 
plot_model(model,show_layer_names=True,show_shapes=True)
```

```python
#now we need to compile the model
model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'],run_eagerly=True)
```

```python
#now lets fit the data to the model 
history=model.fit(x_train,y_train,epochs=10,batch_size=2,verbose=1)
```

```python
model.evaluate(x_test,y_test,batch_size=2,verbose=1)
```

```python
### now lets plot 
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.title('ACCURACY')
plt.xlabel('epoch')
plt.ylabel('accuracy')

plt.show()

plt.plot(history.history['loss'])
plt.title('loss')
plt.xlabel('epoch')
plt.ylabel('loss')

plt.show()

```
