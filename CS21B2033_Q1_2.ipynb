```python
import numpy as np

def calculate_class(weights, bias, feature_vector):
    l=[]
    for i in range(4):
      l.append(feature_vector[i])
    l=np.asarray(l)


    y = np.dot(weights, l) + bias
    return 1 if y >= 0.5 else 0

def error_calculation(weights, bias, feature_vector):
    a = feature_vector[4] - calculate_class(weights, bias, feature_vector)
    b, c, d, e= feature_vector[:4]
    f=1
    return [a * b, a * c, a * d, a * e, a * f]

def updation(weights, bias, feature_vector):
    k = error_calculation(weights, bias, feature_vector)
    current_weights_b=[]
    for i in range(len(weights)):
      current_weights_b.append(weights[i])
    current_weights_b.append(bias)

    updated_weights_b = [0] * len(current_weights_b)

    print("-----------------------------------------------------------------------")

    for i in range(len(k)):
        updated_weights_b[i] = current_weights_b[i] - 0.0001 * k[i]

    print("updated w1 =", updated_weights_b[0])
    print("updated w2 =", updated_weights_b[1])
    print("updated w3 =", updated_weights_b[2])
    print("updated w4 =", updated_weights_b[3])
    print("updated bias =", updated_weights_b[4])

    print("-----------------------------------------------------------------------")



    return updated_weights_b

```

```python
import numpy as np

initial_weight=[0,0,0,0]
initial_bias=0

initial_weight=np.asarray(initial_weight)

#here i will be trianing for a certian number of iterations to ensure convergence
#in other words it is the number of the epochs
for i in range(10):
  print("EPOCHS -> ",i)
  for j in range(len(vectors)):
    final_weight=updation(initial_weight,initial_bias,vectors[j])
    initial_weight=final_weight[:4]
    intial_bias=final_weight[4]








```
